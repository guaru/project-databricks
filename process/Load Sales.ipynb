# Imports

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql import functions as F
```

# Widgets

```python
dbutils.widgets.text("catalog", "project_smartdata")
dbutils.widgets.text("schema_source", "silver")
dbutils.widgets.text("schema_sink", "golden")
```

# Constants

```python
catalog =  dbutils.widgets.get("catalog")
schema_source =  dbutils.widgets.get("schema_source")
schema_sink  =  dbutils.widgets.get("schema_sink")
```

# Read table from silver

```python
df_category_sales =  spark.table(f"{catalog}.{schema_source}.category_sales")
df_product_sales =  spark.table(f"{catalog}.{schema_source}.product_sales")
df_store_sales   =  spark.table(f"{catalog}.{schema_source}.store_sales")
```

# Save table kpi in golden

```python
df_category_sales.write.mode("overwrite").saveAsTable(f"{catalog}.{schema_sink}.kpi_category_sales")
```

```python
df_product_sales.write.mode("overwrite").saveAsTable(f"{catalog}.{schema_sink}.kpi_product_sales")
```

```python
df_store_sales.write.mode("overwrite").saveAsTable(f"{catalog}.{schema_sink}.kpi_store_sales")
```
