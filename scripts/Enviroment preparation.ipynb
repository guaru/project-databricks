```python
%sql
create widget text storageName default "adlsprojectsmartdata";
```

# External location

```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `exlt-data`
URL 'abfss://data@${storageName}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL credential)
COMMENT 'Ubicación de catalogos y tablas'
```

# Catalogs and Schema

```python
%sql
CREATE CATALOG IF NOT EXISTS project_smartdata
MANAGED LOCATION 'abfss://data@${storageName}.dfs.core.windows.net/'
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS project_smartdata.bronze;
CREATE SCHEMA IF NOT EXISTS project_smartdata.silver;
CREATE SCHEMA IF NOT EXISTS project_smartdata.golden;
```
