Permisos esquema y catalogos

```python
%sql
GRANT ALL PRIVILEGES ON CATALOG apple_production TO `DataEnginers`;
GRANT CREATE, USE SCHEMA ON SCHEMA  apple_production.bronze to `DataEnginers`;
GRANT CREATE, USE SCHEMA ON SCHEMA  apple_production.silver to `DataEnginers`;
GRANT CREATE, USE SCHEMA ON SCHEMA  apple_production.golden to `DataEnginers`;

```
